```python
# definir el número de epochs
num_epochs = 100

# definir un valor inicial para la tendencia exponencial
start_value = 0.2

# definir una tasa de crecimiento exponencial
growth_rate = 0.05

# generar una lista de valores exponenciales que se acumulan con el tiempo
trend = start_value * np.exp(np.arange(num_epochs) * growth_rate)

# escalar la lista de valores para que esté dentro del rango de 0 a 1
trend_scaled = minmax_scale(trend, feature_range=(0, 1))

# agregar un término de ruido aleatorio para simular la variabilidad en la precisión
noise = np.random.normal(0, 0.02, num_epochs)

# combinar la tendencia y el ruido para generar la lista de precisión
accuracy = trend_scaled + noise

# graficar la precisión
plt.plot(accuracy)

# agregar etiquetas de eje x e y y título
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy During Training')

# mostrar la gráfica resultante
plt.show()
```

```python
# definir el número de epochs
num_epochs = 200
num_cont = 50

# definir
start_value = 0.2
growth_rate = 0.02
max_value_training = 0.90
max_value_validation = 0.60

# nombre modelo
fig_name = 'B'

cont = []
cont2 = []
for i in range(num_cont):
    c = random.uniform(max_value_training-0.05, max_value_training)
    c2 = random.uniform(max_value_validation-0.03, max_value_validation)
    cont.append(c)
    cont2.append(c2)
cont = np.array(cont)
cont2 = np.array(cont2)


# generar una lista de valores logarítmicos que se acumulan con el tiempo
trend = start_value + growth_rate * np.log(np.arange(1, num_epochs+1))

# escalar la lista de valores para que esté dentro del rango de 0 a 1
trend_scaled = minmax_scale(trend, feature_range=(0, max_value_training))
trend_scaled2 = minmax_scale(trend, feature_range=(0, max_value_validation))

# agregar un término de ruido aleatorio para simular la variabilidad en la precisión
noise = np.random.normal(0, 0.03, num_epochs)
noise2 = np.random.normal(0, 0.008, num_epochs)


# combinar la tendencia y el ruido para generar la lista de precisión
accuracy = np.concatenate((trend_scaled + noise, cont))
accuracy2 = np.concatenate((trend_scaled2 + noise2, cont2))

# graficar la precisión
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(accuracy, label='Training Accuracy')
ax.plot(accuracy2, label='Validation Accuracy')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title('Accuracy of the Model')

# Ajustar los límites y el espacio alrededor del plot
ax.set_xlim(0, num_epochs+num_cont)
ax.set_ylim(0, 1)
fig.tight_layout()

# mostrar la gráfica resultante
plt.savefig(f'images/{fig_name}', bbox_inches='tight')
```
